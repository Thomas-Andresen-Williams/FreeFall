```python
import numpy as np
from scipy.integrate import RK45
import matplotlib.pyplot as plt
ro = 1000 # density [kg/m^3]
C = 0.70 # Drag coefficient [m^2 / s^2]
A = 0.20 # Surface area(feet first) [m^2]
m = 70 # object mass [kg]
g = 9.81 # gravitational acceleration [m/s^2]
iterations = 1000
```

```python
def velocity(t,v):
    return (-g + 0.5 * A * C * v**2 / m ) * t

def distance(t,v):
    return v*t

v0 = np.array([0])
s0 = np.array([0])

solution = RK45(velocity, 0, v0, 100000, max_step
= 0.01, atol = 100000, rtol = 100000, first_step = 0.01)


t_values = []
v_values = []
s_values = []

for i in range(iterations):
    
    solution.step()
    t_values.append(solution.t)
    v_values.append(solution.y[0])
    
    if len(v_values)==1:
        
        s_values.append(v_values[i]*t_values[i])
        
    else:
        
        s_values.append((v_values[i]+v_values[i-1])*(t_values[i] - t_values[i-1])/2 + s_values[i-1])
     
    if solution.status == 'finished':
        break
```

```python
plt.title("Velocity of average human jumping with feet first")
plt.plot(t_values,v_values)
plt.xlabel("time [s]")
plt.ylabel("velocity [m/s]")
plt.show()

plt.title("Corresponding distance")
plt.plot(t_values,s_values)
plt.xlabel("time [s]")
plt.ylabel("distance [m]")
plt.show()

plt.title("For inference")
plt.plot(t_values[0:250],s_values[0:250])
plt.xlabel("time [s]")
plt.ylabel("distance [m]")
plt.show()
```

```python
with open('readme.txt', 'w') as f:
    for i in range(len(t_values[0:250])):
        f.write('t_values[i]')
        f.write('s_values[i]')
        f.write('\n')
```
